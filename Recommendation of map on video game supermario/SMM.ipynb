```python
from keras.layers import Input, Embedding, Dot, Reshape, Dense
from keras.models import Model

print(len(player_index))
print(len(maps_index))

def build_model(embedding_size = 50, classification = False):
    # Both inputs are 1-dimensional
    player = Input(name = 'player', shape = [1])
    map = Input(name = 'map', shape = [1])
    
    # Embedding the book (shape will be (None, 1, 50))
    player_embedding = Embedding(name = 'player_embedding',
                               input_dim = len(player_index),
                               output_dim = embedding_size)(player)
    
    # Embedding the link (shape will be (None, 1, 50))
    map_embedding = Embedding(name = 'map_embedding',
                               input_dim = len(maps_index),
                               output_dim = embedding_size)(map)
    
    # Merge the layers with a dot product along the second axis 
    # (shape will be (None, 1, 1))
    merged = Dot(name = 'dot_product', normalize = True, 
                 axes = 2)([player_embedding, map_embedding])
    
    # Reshape to be a single number (shape will be (None, 1))
    merged = Reshape(target_shape = [1])(merged)
    
    # outputs for classification
    out = Dense(1, activation = 'sigmoid')(merged)
    model = Model(inputs = [player, map], outputs = out)
    
    # Compile using specified optimizer and loss 
    model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', 
                  metrics = ['accuracy'])
    
    return model

model = build_model()
model.summary()
```

```python
#generator
import numpy as np
import random
random.seed(100)
neg_label = 0

def generate_batch(pairs, n_positive = 50, negative_ratio = 1.0):
    """Generate batches of samples for training. 
       Random select positive samples
       from pairs and randomly select negatives."""
    
    # Create empty array to hold batch
    batch_size = n_positive * (1 + negative_ratio)
    batch = np.zeros((batch_size, 3))
    
    # Continue to yield samples
    while True:
        # Randomly choose positive examples
        for idx, (player_id, map_id) in enumerate(random.sample(pairs, n_positive)):
            batch[idx, :] = (player_id, map_id, 1)
        idx += 1
        
        # Add negative examples until reach batch size
        while idx < batch_size:
            
            # Random selection
            random_player = random.randrange(len(list_of_frequent_player))
            random_map = random.randrange(len(list_of_frequent_maps))
            
            # Check to make sure this is not a positive example
            if (random_player, random_map) not in pairs:
                
                # Add to batch and increment index
                batch[idx, :] = (random_player, random_map, neg_label)
                idx += 1
                
        # Make sure to shuffle order
        np.random.shuffle(batch)
        yield {'player': batch[:, 0], 'map': batch[:, 1]}, batch[:, 2]
```

```python
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
n_positive = 1024

gen = generate_batch(like_pairs, n_positive, negative_ratio = 2)

# Train
h = model.fit_generator(gen, epochs = 15, 
                        steps_per_epoch = (len(like_pairs) // n_positive))
```
